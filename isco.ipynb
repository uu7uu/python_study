```python
import numpy as np
from astropy.io import fits
import pandas as pd
import matplotlib.pyplot as plt
```

```python
a1=np.arange(0,1.01,0.01)
a0=np.arange(-1,0,0.01)
print(a1)
```

```python
a=a1
z1= 1+((1-a**2)**(1/3))*((1+a)**(1/3)+(1-a)**(1/3))
z2=(3*a**2+z1**2)**(1/2)
isco=3+z2-((3-z1)*(3+z1+2*z2))**(1/2) #a>0
#rin=3+z2+((3-z1)*(3+z1+2*z2))**(1/2) a<0
print(isco)
```

```python
a=a0
z1= 1+((1-a**2)**(1/3))*((1+a)**(1/3)+(1-a)**(1/3))
z2=(3*a**2+z1**2)**(1/2)
#isco=3+z2-((3-z1)*(3+z1+2*z2))**(1/2) #a>0
rin=3+z2+((3-z1)*(3+z1+2*z2))**(1/2) #a<0
print(rin)
```

```python
plt.figure(dpi=150)
plt.plot(a0,rin,linewidth=1.2,color='salmon') #,
plt.plot(a1,isco,linewidth=1.2,color='royalblue')
# plt.title("Squares Numbers",fontsize=24) #
plt.xlabel(r"$\mathrm{a^{*}=a/M}$",fontsize=20) #
plt.ylabel(r"$\mathrm{R_{ISCO}/r_{g}}$",fontsize=20)
plt.tick_params(axis='both',labelsize=14) #
plt.xlim(-1.0,1.0)
plt.ylim(0,10.0)
plt.text(-0.3, 8, r"Retrograde", fontsize=12, color='salmon',verticalalignment='center',horizontalalignment='right',rotation=-20)
plt.text(0.7, 4.6, r"Prograde", fontsize=12, color='royalblue',verticalalignment='center',horizontalalignment='right',rotation=-22)
plt.text(0.9, 6.3, r"Schwarzchild", fontsize=12, color='slategrey',verticalalignment='center',horizontalalignment='right',rotation=0)
plt.text(0.9, 1.3, r"Extreme Kerr", fontsize=12, color='slategrey',verticalalignment='center',horizontalalignment='right',rotation=0)

plt.axhline(y=6,color='slategrey',linewidth=1,linestyle='--')
plt.axhline(y=1,color='slategrey',linewidth=1,linestyle='--')
plt.show()
plt.savefig('11.png')
```
